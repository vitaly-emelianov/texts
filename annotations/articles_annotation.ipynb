```python
with open("articles.txt", 'r') as f:
    data_in = f.read()
articles = data_in.split('@')

print(articles[0])
print(len(articles))
```

```python
test_article = articles[0].strip()
sentences = test_article.split('.')

from sklearn.feature_extraction.text import CountVectorizer

count_vect = CountVectorizer(analyzer='word')
sentences_counts = count_vect.fit_transform(sentences)
article_counts = count_vect.fit_transform([test_article])
print(sentences_counts.shape)
```

```python
from sklearn.metrics.pairwise import cosine_similarity
import numpy as np

similarities = cosine_similarity(sentences_counts, article_counts)
print(similarities)
```
